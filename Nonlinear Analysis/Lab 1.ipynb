```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

# Input functions

```python
def P(x,y):
    return -(6 + x - y**2)

def Q(x,y):
    return x * (y - x)
```

```python
def V_dx(x,y):
    return P(x,y)/math.hypot(P(x,y),Q(x,y))

def V_dy(x,y):
    return Q(x,y)/math.hypot(P(x,y),Q(x,y))
```

# Phase portrait with standard alpha

```python
N_iter = 1000
alpha_x = alpha_y = 10 **(-2)
start_x = -5
start_y = -5
end_x = 5
end_y = 5

for i in np.linspace(start_x, end_x, num=20):
    for j in np.linspace(start_y, end_y, num=20):
        x = [i]
        y = [j]

        for k in range(1, N_iter):
            if math.hypot(P(x[k-1], y[k-1]), Q(x[k-1], y[k-1])) == 0:
                break
            x.append(x[k-1] + alpha_x * V_dx(x[k-1],y[k-1]))
            y.append(y[k-1] + alpha_y * V_dy(x[k-1],y[k-1]))

        plt.axis([start_x, end_x, start_y, end_y])
        plt.plot(x,y)
```

# Phase portraits with dynamic alpha

```python
N_iter = 100
alpha_x = alpha_y = 10 **(-1)
start_x = -5
start_y = -5
end_x = 5
end_y = 5

for i in np.linspace(start_x, end_x, num=20):
    for j in np.linspace(start_y, end_y, num=20):
        x = [i]
        y = [j]

        for k in range(1, N_iter):
            if math.hypot(P(x[k-1], y[k-1]), Q(x[k-1], y[k-1])) == 0:
                break
            x.append(x[k-1] + alpha_x * V_dx(x[k-1],y[k-1]))
            y.append(y[k-1] + alpha_y * V_dy(x[k-1],y[k-1]))

        plt.axis([start_x, end_x, start_y, end_y])
        plt.plot(x,y)
```

```python
N_iter = 700
base_alpha_x = base_alpha_y = 10 **(-2)
start_x = -5
start_y = -5
end_x = 5
end_y = 5

for i in np.linspace(start_x, end_x, num=20):
    for j in np.linspace(start_y, end_y, num=20):
        x = [i]
        y = [j]

        alpha_x = []
        alpha_y = []

        for k in range(1, N_iter):
            alpha_x.append(base_alpha_x)
            alpha_y.append(base_alpha_y)

            if math.hypot(P(x[k-1], y[k-1]),Q(x[k-1], y[k-1])) == 0:
                break
            x.append(x[k-1] + alpha_x[-1] * V_dx(x[k-1],y[k-1]))
            y.append(y[k-1] + alpha_y[-1] * V_dy(x[k-1],y[k-1]))

            if math.hypot(x[k-2], x[k-3]) == 0 or math.hypot(y[k-2], y[k-3]) == 0:
                break
            alpha_x.append(alpha_x[k-1]*math.hypot(x[k-1], x[k-2])/math.hypot(x[k-2], x[k-3]))
            alpha_y.append(alpha_y[k-1]*math.hypot(y[k-1], y[k-2])/math.hypot(y[k-2], y[k-3]))

        plt.axis([start_x, end_x, start_y, end_y])
        plt.plot(x,y)
```
